Scenario 6 : A grading system determines whether a student has passed or failed based on their score. Write logic to check if a student has passed a subject by scoring at least 40 marks.

Logic Written Below :

1) Get the inputs from the user for the variable marks.
2) Based on entered marks, to check the condition using if-else.
3) If marks greater than or equal to 40 students passed in examination
4) If marks lesser than 40 students failed in examination

```python
# Method - 1 : Hardcoding the marks 40 and compare the values using if-else
marks = 40
if marks >= 40:
    print("Student has passed a subject")
else:
    print("Student has failed a subject and minimum score at least 40 marks")
```

```python
# Method - 1a : Hardcoding the marks 40 and compare the values using if-else
marks = 39
if marks >= 40:
    print("Student has passed a subject")
else:
    print("Student has failed a subject and minimum score at least 40 marks")
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
marks = int(input("Enter the Student Marks : "))
if marks >= 40:
    print("Student has passed a subject")
else:
    print("Student has failed a subject and minimum score at least 40 marks")
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def studentMarks():
    marks = int(input("Enter the Student Marks : "))
    if marks >= 40:
        print("Student has passed a subject")
    else:
        print("Student has failed a subject and minimum score at least 40 marks")
studentMarks()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def studentMarks(marks):
    if marks >= 40:
        return "Student has passed a subject"
    else:
        return "Student has failed a subject and minimum score at least 40 marks"
studentMarks(40)
```

```python
# Method - 5 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class Marks:
    def studentMarks():
        marks = int(input("Enter the Student Marks : "))
        if marks >= 40:
            print("Student has passed a subject")
        else:
            print("Student has failed a subject and minimum score at least 40 marks")
Marks.studentMarks()
```
