Scenario 4 : In a secure login system, certain features are enabled only for users with prime-numbered user IDs. Write logic to check if a given number is prime.

Logic Written Below.

1) Get the input from the user for the variable user_id.
2) If the user_id inputs divisible by 1 and the number itself no other divisiors, the prime users enable access
3) If the user_id inputs not divisible by 1 and divide by some other numbers than the prime users features not enable.

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
user_id = int(input("Enter the Number : "))
if user_id > 1:
    for i in range(2, user_id):
        if user_id % i == 0:
            print("Prime Users Features not Enabled")
            break
    else:
        print("Prime Users Features Enabled")
else:
    print("Prime Users Features not Enabled")
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def PrimeCheck():
    user_id = int(input("Enter the Number : "))
    if user_id > 1:
        for i in range(2, user_id):
            if user_id % i == 0:
                print("Prime Users Features not Enabled")
                break
        else:
            print("Prime Users Features Enabled")
    else:
        print("Prime Users Features not Enabled")
PrimeCheck()
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def PrimeCheck(user_id):
    if user_id > 1:
            for i in range(2, user_id):
                if user_id % i == 0:
                    return "Prime Users Features not Enabled"
                    break
            else:
                return "Prime Users Features Enabled"
    else:
        return "Prime Users Features not Enabled"
PrimeCheck(11)
```
