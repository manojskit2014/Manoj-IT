Scenario 3 : A company provides employees with a 10% bonus if their salary exceeds $50,000. Write logic to determine the bonus amount based on the given salary.

Logic Written Below :
1) Get the user inputs from the salary and percentage
2) Inititalize one variable total_amount to hold the based on employee salary exceeds 50,000 and bonus 10%
3) Use if condition to check, if the salary exceeds 50,000 it will print the salary amount with bonus 10%
4) If the amount not exceeds 50,000 it will print the salary amount without bonus 10%

```python
# Method - 1 : Hardcoding the values 50000 and comparison values using if
salary_1 = 50000
percentage = 10/100
amount_1 = salary_1 + (salary_1 * percentage)
if salary_1 >= 50000:
    print("Your Salary with Bonus : ", amount_1)
```

```python
# Method - 1a : Hardcoding the values 30000 and comparison values using if
salary_2 = 30000
percentage = 10/100
amount_2 = salary_2 + (salary_2 * percentage)
if salary_2 < 50000:
    print("Your Salary without Bonus : ", amount_2)
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
salary = float(input("Enter Your Salary : "))
percentage = float(input("Enter Your Percentage : "))
amount = salary + (salary * percentage / 100)
if amount >= 50000:
    print("Your Salary with Bonus : ", amount)
else:
    print("Your Salary without Bonus : ", amount)
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def employeeSalary():
    salary = float(input("Enter Your Salary : "))
    percentage = float(input("Enter Your Percentage : "))
    amount = salary + (salary * percentage / 100)
    if amount >= 50000:
        print("Your Salary with Bonus : ", amount)
    else:
        print("Your Salary without Bonus : ", amount)
employeeSalary()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def employeeSalary(salary, percentage):
    amount = salary + (salary * percentage / 100)
    if amount >= 50000:
        return "Your Salary with Bonus : ", amount
    else:
        return "Your Salary without Bonus : ", amount
employeeSalary(50000, 10)
```

```python
# Method - 5 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class EmployeeSalaryBonus:
    def employeeSalary():
        salary = float(input("Enter Your Salary : "))
        percentage = float(input("Enter Your Percentage : "))
        amount = salary + (salary * percentage / 100)
        if amount >= 50000:
            print("Your Salary with Bonus : ", amount)
        else:
            print("Your Salary without Bonus : ", amount)
EmployeeSalaryBonus.employeeSalary()
```

```python
# Defining with class and functions using return statement
class Employees:
    def employeeBonus():
        sal = float(input("Enter Your Salary :"))
        per = float(input("Enter Your Percentage :"))
        amount = sal + (sal * per / 100) # Percentage Formula
        if amount >= 50000:
            print("Your Salary with added Bonus : ", amount)
        else:
            print("Your Salary with not added Bonus : ", amount)
        return amount
Employees.employeeBonus()
```
