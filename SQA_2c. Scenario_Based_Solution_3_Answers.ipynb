Scenario 3 : A mobile payment app uses a simple checksum validation where reversing a transaction ID helps detect fraud. Write logic to take a number and return its reverse.

Logic Written Below :

1) Get the transaction ID input from the user.
2) Convert it into string.
3) Reverse the string.
4) convert it from reverse string to int
5) print the reversed number

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
trans_id = int(input("Enter the Transaction ID : "))
print("The Transaction ID is : ", trans_id)
print(type(trans_id))
# Converting int to str
st = str(trans_id)
print(type(st))
# Reversing the string using slice operator
reverse_st = st[::-1]
print(reverse_st)
# Again reversing the string to int
reverse_int = int(reverse_st)
print(type(reverse_int))
print("The Reversed Transcation ID is : ", reverse_int)
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def checksum(trans_id):
    print(type(trans_id))
    # Converting int to str
    st = str(trans_id)
    print(type(st))
    # Reversing the string using slice operator
    reverse_st = st[::-1]
    print(reverse_st)
    # Again reversing the string to int
    reverse_int = int(reverse_st)
    print(type(reverse_int))
    return "The Reversed Transcation ID is : ", reverse_int
checksum(123456)
```
