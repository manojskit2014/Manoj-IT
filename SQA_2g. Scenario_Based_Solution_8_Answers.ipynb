Scenario 8 : A banking system processes withdrawal requests and ensures the user has enough balance.

Logic Written Below.

1) Get the input from the user for the balance amount.
2) if withdrawal amount less than or equal to balance amount, user able to withdrawal the amount
3) And update the balance amount from withdrawal amount.
4) if withdrawal amount exceeds than balance amount, insufficient funds message should display in console.

```python
# Method - 1 : Hardcoding the balance amount and withdrawal the amount less than balance amount
bal_amount = 30000
withdrawal_amt = 10000
if withdrawal_amt <= bal_amount:
    original_balance = bal_amount - withdrawal_amt
    print("User able to Withdraw the amount")
    print("After withdrawal the balance amount is : ", original_balance)
else:
    print("Insufficient Funds")
```

```python
# Method - 1a : Hardcoding the balance amount and withdrawal the amount exceeds than balance amount
bal_amount = 30000
withdrawal_amt = 40000
if withdrawal_amt <= bal_amount:
    original_balance = bal_amount - withdrawal_amt
    print("User able to Withdraw the amount")
    print("After withdrawal the balance amount is : ", original_balance)
else:
    print("Insufficient Funds")
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
bal_amount = int(input("Enter the Balance Amount : "))
withdrawal_amount = int(input("Enter the Withdrawal Amount : "))
if withdrawal_amount <= bal_amount:
    original_balance = bal_amount - withdrawal_amount
    print("User able to Withdraw the amount")
    print("After withdrawal the balance amount is : ", original_balance)
else:
    print("Insufficient Funds")
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def balanceAmount():
    bal_amount = int(input("Enter the Balance Amount : "))
    withdrawal_amt = int(input("Enter the Withdrawal Amount : "))
    if withdrawal_amt <= bal_amount:
        original_balance = bal_amount - withdrawal_amt
        print("User able to Withdraw the amount")
        print("After withdrawal the balance amount is : ", original_balance)
    else:
        print("Insufficient Funds")
balanceAmount()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def balanceAmount(bal_amount, withdrawal_amt):
    if withdrawal_amt <= bal_amount:
        original_balance = bal_amount - withdrawal_amt
        return '''User able to Withdraw the amount, After withdrawal the balance amount is : ''', original_balance 
    else:
        return "Insufficient Funds"
balanceAmount(1000, 500)
```
