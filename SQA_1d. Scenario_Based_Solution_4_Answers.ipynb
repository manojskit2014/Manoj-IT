4.	Scenario: A program evaluates a number to determine if it is even or odd. Write logic to check whether a given number is even or odd.

Logic Written Below :
1) Get the input from the user and assign it to one variable x.
2) Using if condition x % 2 == 0, the given number is even
3) x % 2 !=0, the given number is odd.

```python
# Method - 1 : Hardcoding the values 50000 and comparison values using if
x = 4
if x % 2 == 0:
    print("The given number is even ")
```

```python
# Method - 1a : Hardcoding the values 30000 and comparison values using if
x = 5
if x % 2 != 0:
    print("The given number is odd ")
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
x = int(input("Enter the Number : "))
if x % 2 == 0:
    print("The given number is even")
else:
    print("The given number is odd")
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def oddeven():
    x = int(input("Enter the Number : "))
    if x % 2 == 0:
        print("The given number is even")
    else:
        print("The given number is odd")
oddeven()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def oddeven(x):
    if x % 2 == 0:
        return "The given number is even"
    else:
        return "The given number is odd"
oddeven(6)
```

```python
oddeven(7)
```

```python
# Method - 5 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class OddorEven:
    def oddeven():
        x = int(input("Enter the Number : "))
        if x % 2 == 0:
            print("The given number is even")
        else:
            print("The given number is odd")
OddorEven.oddeven()
```
