Scenario 10 : A program filters out only even numbers from a given list. Write logic to extract and return only the even numbers from a list.

Logic Written Below.
1) Get the list of numbers from the user input.
2) And iterate all the numbers from the list through for() loop and print all the numbers.
3) If the number is modulo by 2 is equals 0, extract and return from the list.

```python
# Method - 1 : Hardcoding the list of numbers and return only even numbers from the list
list1 = [19,20,17,18,16,15,14,13,11,12,9,10,8,7,5,4,6,2,3,1]
print("Before Sorting : ", list1)
list1.sort()
print("After Sorting : ", list1)
list2 = []
for each_element in list1:
    if each_element % 2 == 0:
        list2.append(each_element)
print("The final even numbers from the list2 is : ", list2)
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
list1 = eval(input("Enter the List of Numbers : "))
print("Before Sorting : ", list1)
list1.sort()
print("After Sorting : ", list1)
list2 = []
for each_element in list1:
    if each_element % 2 == 0:
        list2.append(each_element)
print("The final even numbers from the list2 is : ", list2)
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def returnListEven():
    list1 = eval(input("Enter the List of Numbers : "))
    print("Before Sorting : ", list1)
    list1.sort()
    print("After Sorting : ", list1)
    list2 = []
    for each_element in list1:
        if each_element % 2 == 0:
            list2.append(each_element)
    print("The final even numbers from the list2 is : ", list2))
returnListEven()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def returnListEven(list1):
    print("Before Sorting : ", list1)
    list1.sort()
    print("After Sorting : ", list1)
    list2 = []
    for each_element in list1:
        if each_element % 2 == 0:
            list2.append(each_element)
    return "The final even numbers from the list2 is : ", list2
returnListEven([2,5,3,4])
```

```python
# Method - 5 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class ReturnListEven:
    def returnListEven(list1):
        print("Before Sorting : ", list1)
        list1.sort()
        print("After Sorting : ", list1)
        list2 = []
        for each_element in list1:
            if each_element % 2 == 0:
                list2.append(each_element)
        return "The final even numbers from the list2 is : ", list2
ReturnListEven.returnListEven([2,5,3,4])
```
