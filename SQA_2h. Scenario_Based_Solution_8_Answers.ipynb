Scenario 8 : A low-level networking application requires decimal numbers to be converted into binary format before transmission. Write logic to convert a given decimal number into its binary equivalent.

Logic Written Below :

1) Get the decimal number n from the user.
2) Create an empty string binary to store the binary digits.
3) Loop: While n > 0:
4) Find the remainder when n is divided by 2 rem = n % 2.
5) Update n = n // 2 (integer division).
6) If input is 0, directly set binary = "0".
7) Print the final binary string.

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
n = int(input("Enter a Decimal Number : "))
binary = ""
if n == 0:
    binary = "0"
else:
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary   # prepend remainder
        n = n // 2
print("Binary Equivalent is :", binary)
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def decimaltobinary():
    n = int(input("Enter a Decimal Number : "))
    binary = ""
    if n == 0:
        binary = "0"
    else:
        while n > 0:
            remainder = n % 2
            binary = str(remainder) + binary   # prepend remainder
            n = n // 2
    print("Binary Equivalent is :", binary)
decimaltobinary()
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def decimaltobinary(n):
    binary = ""
    if n == 0:
        binary = "0"
    else:
        while n > 0:
            remainder = n % 2
            binary = str(remainder) + binary   # prepend remainder
            n = n // 2
    return "Binary Equivalent is :", binary
decimaltobinary(10)
```
