Scenario 1 : You are developing a banking application that categorizes transactions based on the amount entered.
Write logic to determine whether the amount is positive, negative, or zero.

Logic Written Below :

1) Get the transcations amount input from the user.
2) Based on the amount entered by user, Using if-elif-else conditon to validate the positive, negative or zero.
3) If the user enters the amount 100, it sayas the given amount is positive.
4) If the user enters the amount below 0, it says the given amount is negative.
5) If the user enters the amount 0, it says the given amount is zero.

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
transactions = int(input("Enter the Amount: "))
if transactions >= 1:
    print("The entered amount is positive")
elif transactions < 0:
    print("The entered amount is negative")
else:
    print("The entered amount is zero")
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def transactions():
    transactions = int(input("Enter the Amount: "))
    if transactions >= 1:
        print("The entered amount is positive")
    elif transactions < 0:
        print("The entered amount is negative")
    else:
        print("The entered amount is zero")
transactions()
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def transactions(amount):
    if amount >= 1:
        return "The entered amount is positive"
    elif amount < 0:
        return "The entered amount is negative"
    else:
        return "The entered amount is zero"
transactions(1)
```

```python
# Method - 4 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class BankApplications:
    def transactions():
        transactions = int(input("Enter the Amount: "))
        if transactions >= 1:
            print("The entered amount is positive")
        elif transactions < 0:
            print("The entered amount is negative")
        else:
            print("The entered amount is zero")
BankApplications.transactions()
```
