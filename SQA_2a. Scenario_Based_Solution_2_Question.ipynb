Scenario 2 : A program processes a list of numbers and needs to find the largest value. Write logic to identify and return the largest number from a given list.

Logic Written Below :
1) Get the user inputs from list of numbers.
2) Initialize a variable to hold the largest list of numbers.
3) Iterate list of numbers through for loop
4) Compare each and every number with a variable to hold the list of numbers.
5) If larger number is found, return the largest number from the list of numbers.

```python
# Method - 1 : Hardcoding the list and comparison values
list = [1,2,4,20,10,13,5]
for i in list:
    if i == 20:
        print("The largest number in the given list is : ", i)
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
list = eval(input("Enter Your List : "))
print("Before Sorting : ", list)
list.sort()
print("After Sorting : ", list)
largest = list[-1]
for each_element in list:
    print(each_element)
    if each_element == largest:
        print("The largest number in the given list is : ", each_element)
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def largestNumbers():
    list = eval(input("Enter Your List : "))
    print("Before Sorting : ", list)
    list.sort()
    print("After Sorting : ", list)
    largest = list[-1]
    for each_element in list:
        print(each_element)
        if each_element == largest:
            print("The largest number in the given list is : ", each_element)
largestNumbers()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def largestNumbers(list):
    print("Before Sorting : ", list)
    list.sort()
    print("After Sorting : ", list)
    for each_element in list:
        print(each_element)
        largest = list[-1]
        if each_element == largest:
            return "The largest number from the list : ", each_element
largestNumbers([1,2,4,20,10,13,5])
```

```python
# Method - 5 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class LargestNumberList:
    def largestNumbers():
        list = eval(input("Enter Your List : "))
        print("Before Sorting : ", list)
        list.sort()
        print("After Sorting : ", list)
        largest = list[-1]
        for each_element in list:
            print(each_element)
            if each_element == largest:
                print("The largest number in the given list is : ", each_element)
LargestNumberList.largestNumbers()
```

Scenario: A company provides employees with a 10% bonus if their salary exceeds $50,000.
Write logic to determine the bonus amount based on the given salary.

```python
# Defining with class and functions using return statement
class Employees:
    def employeeBonus():
        sal = float(input("Enter Your Salary :"))
        per = float(input("Enter Your Percentage :"))
        amount = sal + (sal * per / 100) # Percentage Formula
        if sal >= 50000:
            print("Your Salary with added Bonus : ", amount)
        else:
            print("Your Salary with not added Bonus : ", amount)
        return amount
Employees.employeeBonus()
```
