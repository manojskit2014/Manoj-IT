Scenario 7 : A password manager needs to strengthen weak passwords by swapping the first and last characters of user-generated passwords. Write logic to perform this operation on a given string.

Logic Written Below :

1) Get the string from the user inputs.
2) If string is less than or equal to 2 characters, print as it is.
3) If string is more than or equal to 2 characters swap the first and last characters and keep middle part as it is
4) And print the swapped string in console.

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
strings = input("Enter the String : ")
if len(strings) <= 2:
    print("The string is only 2 characters : ", strings)
else:
    swapped = strings[-1] + strings[1:-1] + strings[0]
    print("Swapped String is : ", swapped)
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def swapping():
    strings = input("Enter the String : ")
    if len(strings) <= 2:
        print("The string is only 2 characters : ", strings)
    else:
        swapped = strings[-1] + strings[1:-1] + strings[0]
        print("Swapped String is : ", swapped)
swapping()
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def swapping(strings):
    if len(strings) <= 2:
        print("The string is only 2 characters : ", strings)
    else:
        swapped = strings[-1] + strings[1:-1] + strings[0]
        print("Swapped String is : ", swapped)
swapping("password")
```
