Scenario 1 : A system checks if a user is eligible to vote based on their age.
Write logic to ask the user for their age and determine if they are eligible to vote based on whether they are 18 or older.

Logic Written Below :
1) Get the input from the user.
2) After input from the user, It checks the condition.
3) If condition is satisified which means condition >= 18 eligible for vote.
4) If condition is not satisified, which means condition < 18 not eligibel for vote.

```python
# Method - 1 : Hardcoding the value and written the code.
age = 18
if age >= 18:
    print("Eligible For Vote")
else:
    print("Not Eligible For Vote")
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
age = int(input("Enter Your Age : "))
if age >= 18:
    print("Eligible For Vote")
else:
    print("Not Eligible For Vote")
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def eligible():
    age = int(input("Enter Your Age : "))
    if age >= 18:
        print("Eligible For Vote")
    else:
        print("Not Eligible For Vote")
eligible()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the use and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def eligible(age):
    if age >= 18:
        return "Eligible For Vote"
    else:
        return "Not Eligible For Vote"
eligible(17)
```

```python
eligible(18)
```

```python
# Method - 5 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class Age:
    def eligible():
        age = int(input("Enter Your Age : "))
        if age >= 18:
            print("Eligible For Vote")
        else:
            print("Not Eligible For Vote")
Age.eligible()
```

Scenario 2 : A program processes a list of numbers and needs to find the largest value.
Write logic to identify and return the largest number from a given list.

```python
# without functions, Here hardcoded the list and comparison values
list = [1,2,4,20,10,13,5]
for i in list:
    if i == 20:
        print("The largest number in the given list is : ", i)
```

```python
# without functions, getting the input from the user and not hardcoded the list and comparison values but used inbuilt funcitons max() and used for()
# and if() loops
list = eval(input("Enter the Number :")) # 
print("After Sorting List : ")
list.sort()
for each_element in list:
    print(each_element)
    if each_element == max(list):
        print("The largest number in the given list is : ", each_element)
```

```python
# without functions, getting the input from the user and not hardcoded the list and comparison values but didn't use inbuilt funcitons max()
# and didn't use for() and if() loops
list = eval(input("Enter the Number : ")) # 
print("Before Sorting List : ", list)
list.sort()
print("After Sorting List : ", list)
largest = list[-1]
print("The largest number in the given list is : ", largest)
```

```python
# Using Functions to find the largest number in the given list.
def largestNumber():
    list = eval(input("Enter the Number :")) # getting the input from the user and not hardcoded any value without functions
    print("After Sorting List : ")
    list.sort()
    for each_element in list:
        print(each_element)
        if each_element == max(list):
            print("The largest number in the given list is : ", each_element)
            return each_element
largestNumber()
```

```python
# Defining with class
class largestLM:
    def largestNumber():
        list = eval(input("Enter the Number : ")) # 
        print("Before Sorting List : ", list)
        list.sort()
        print("After Sorting List : ", list)
        largest = list[-1]
        print("The largest number in the given list is : ", largest)
largestLM.largestNumber()
```

Scenario: A company provides employees with a 10% bonus if their salary exceeds $50,000.
Write logic to determine the bonus amount based on the given salary.

```python
# Defining with class and functions using return statement
class Employees:
    def employeeBonus():
        sal = float(input("Enter Your Salary :"))
        per = float(input("Enter Your Percentage :"))
        amount = sal + (sal * per / 100) # Percentage Formula
        if sal >= 50000:
            print("Your Salary with added Bonus : ", amount)
        else:
            print("Your Salary with not added Bonus : ", amount)
        return amount
Employees.employeeBonus()
```
