Create a class and function, and list out the items in the list
SubfieldsInAI.Subfields()
Sub-fields in AI are:
Machine Learning
Neural Networks
Vision
Robotics
Speech Processing
Natural Language Processing

```python
# Without Function written the code.
lists = ['Machine Learning', 'Neural Networks', 'Vision', 'Robotics', 'Speech Processing', 'Natural Language Processing']
print("Sub-fields in AI are : ")
for subfieldsInAI in lists:
    print(subfieldsInAI)
```

```python
# With Function written the code and after calling the function name (Function Invocation).
def subFields():
    lists = ['Machine Learning', 'Neural Networks', 'Vision', 'Robotics', 'Speech Processing', 'Natural Language Processing']
    print("Sub-fields in AI are : ")
    for subfieldsInAI in lists:
        print(subfieldsInAI)
subFields() # Calling the function name
```

```python
# With Function written the code and after calling the function name (Function Invocation). Storting with one variable and am printing the 
# value of the variable what stored with using return statement
num = int(input("Enter the Number : "))
def oddEven():
    if num % 2 == 0:
        print("The given number is an Even Number")
        msg = "Even Number"
    else:
        print("The given number is an Odd Number")
        msg = "Odd Number"
    return msg
msg = oddEven()
msg   
```
