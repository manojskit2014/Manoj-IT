Scenario 6 : A unique lottery system assigns ticket numbers where only Armstrong numbers win the jackpot. Write logic to check whether a given number is an Armstrong number.

Logic Written Below :

1) Get the inputs from the user as stored in one variable x.
2) And convert int to string
3) And store that string into one variable named as digits
4) Assign one variable sum = 0
5) Implement the formula for armstrong sum = sum + (digit ^ number_of_digits).
6) if x == sum, print armstrong number
7) Else x != sum, print not armstrong number

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
n = int(input("Enter the Number : "))
n_str = str(n)
n_digits = len(n_str)
sum = 0
for each_digit in n_str:
    sum = sum + int(each_digit) ** n_digits
if sum == n:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def armstrong():
    n = int(input("Enter the Number : "))
    n_str = str(n)
    n_digits = len(n_str)
    sum = 0
    for each_digit in n_str:
        sum = sum + int(each_digit) ** n_digits
    if sum == n:
        print("Armstrong Number")
    else:
        print("Not an Armstrong Number")
armstrong()
```
