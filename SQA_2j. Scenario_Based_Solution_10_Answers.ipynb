Scenario 10 : A plagiarism detection tool compares words from different documents and checks if they are anagrams (same characters but different order). Write logic to check whether two given strings are anagrams.

Logic Written Below

1) Get two input strings (str1, str2).
2) Convert both to lowercase.
3) If lengths differ not anagrams.
4) Else, sort both strings:
    If sorted results are identical they are anagrams.
    Otherwise not anagrams.

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code. Decimal to Binary Conversion
str_1 = input("Enter the String1 : ").lower()
str_2 = input("Enter the String2 : ").lower()
if len(str_1) == len(str_2) and sorted(str_1) == sorted(str_2):
    print("They are Anagrams")
else:
    print("They are not Anagrams")
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def anagrams():
    str_1 = input("Enter the String1 : ").lower()
    str_2 = input("Enter the String2 : ").lower()
    if len(str_1) == len(str_2) and sorted(str_1) == sorted(str_2):
        print("They are Anagrams")
    else:
        print("They are not Anagrams")
anagrams()
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def anagrams(str_1, str_2):
    if len(str_1) == len(str_2) and sorted(str_1) == sorted(str_2):
        return "They are Anagrams"
    else:
        return "They are not Anagrams"
anagrams("listen", "silent")
```

```python
# Method - 4 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class ANAGMS:
    def anagrams():
        str_1 = input("Enter the String1 : ").lower()
        str_2 = input("Enter the String2 : ").lower()
        if len(str_1) == len(str_2) and sorted(str_1) == sorted(str_2):
            print("They are Anagrams")
        else:
            print("They are not Anagrams")
ANAGMS.anagrams()
```
