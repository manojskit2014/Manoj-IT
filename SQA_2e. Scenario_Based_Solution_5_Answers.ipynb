Scenario: A scientist is working on permutations and needs to calculate the factorial of numbers frequently. Write logic to find the factorial of a given number using recursion.

Logic Written Below :

1) Get the input from the user.
2) Based on input, If n==0 or n==1 return 1
3) Else, return n * fact(n-1) calculate factorial and print the output in console

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
n = int(input("Enter the Number : "))
f = 1
for i in range(1, n + 1):
    f = f * i
print("The Factorial of {} is : {}".format(n,f))
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def factorial():
    n = int(input("Enter the Number : "))
    f = 1
    for i in range(1, n + 1):
        f = f * i
    print("The Factorial of {} is : {}".format(n,f))
factorial()
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def factorial(n):
    f = 1
    for i in range(1, n + 1):
        f = f * i
    return "The Factorial of {} is : {}".format(n,f)
factorial(5)
```

```python
# Method - 4 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class FactorialNumber:
    def factorial():
        n = int(input("Enter the Number : "))
        f = 1
        for i in range(1, n + 1):
            f = f * i
        print("The Factorial of {} is : {}".format(n,f))
FactorialNumber.factorial()
```

```python
# Method - 5 : Using Recursion
def factorial(n):
    print("Calling factorial({})".format(n))   # print before recursion
    if n == 0 or n == 1:
        print("Base case reached with {} = ".format(n))
        return 1
    else:
        result = n * factorial(n - 1)
        print("Returning {} * factorial({}) = {}".format(n, n-1, result))  # print after recursion
        return result

# Example usage
print("Final Result:", factorial(5))
```
