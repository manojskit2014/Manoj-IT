Scenario 9 : A calendar system verifies whether a given year is a leap year based on standard leap year rules.

Logic Written Below :
1) Read the input from the user for the year.
2) If the year is divisible by 400 than it is a leap year
3) If the year is divisible by 100 than it is not a leap year.
4) If the year is divisible by 4 than it is a leap year.

```python
# Method - 1 : Hardcoding the year and check the leap year or not leap year for 1990
leap_year = 2026
if leap_year % 400 == 0:
    print(leap_year, "is a leap year")
elif leap_year % 100 == 0:
    print(leap_year, "is not a leap year")
elif leap_year % 4 == 0:
    print(leap_year, "is a leap year")
else:
    print(leap_year, "is not a leap year")
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
leap_year = int(input("Enter the Year : "))
if leap_year % 400 == 0:
    print(leap_year, "is a leap year")
elif leap_year % 100 == 0:
    print(leap_year, "is not a leap year")
elif leap_year % 4 == 0:
    print(leap_year, "is a leap year")
else:
    print(leap_year, "is not a leap year")
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def Leap_Year():
    leap_year = int(input("Enter the Year : "))
    if leap_year % 400 == 0:
        print(leap_year, "is a leap year")
    elif leap_year % 100 == 0:
        print(leap_year, "is not a leap year")
    elif leap_year % 4 == 0:
        print(leap_year, "is a leap year")
    else:
        print(leap_year, "is not a leap year")
Leap_Year()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def Leap_Year(leap_year):
    if leap_year % 400 == 0:
        return leap_year, "is a leap year"
    elif leap_year % 100 == 0:
        return leap_year, "is not a leap year"
    elif leap_year % 4 == 0:
        return leap_year, "is a leap year"
    else:
        return leap_year, "is not a leap year"
Leap_Year(1990)
```
