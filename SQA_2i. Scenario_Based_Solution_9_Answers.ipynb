Scenario 9 : A text-processing tool helps summarize articles by identifying the most significant words. Write logic to find the longest word in a sentence.

Logic Written Below :

1) Get the sentence inputs from the user and store it in one variable 'sentence'
2) Split the sentence into individual words.
3) Initialize a variable longest_word = "".
4) Loop through each word in the list:
5) If the length of the current word is greater than the length of longest_word, update longest_word = current_word.
6) After the loop ends, print longest_word.

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code. Decimal to Binary Conversion
sentence = input("Enter a Decimal Number : ")
individual_word = sentence.split()
print(individual_word)
longest_word = ""
for each_element in individual_word:
    if len(each_element) > len(longest_word):
        longest_word = each_element
print("The Longest Word is : ", longest_word)
```
