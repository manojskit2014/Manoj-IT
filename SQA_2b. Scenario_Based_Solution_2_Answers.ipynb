Scenario 2 : A digital locker requires users to enter a numerical passcode. As part of a security feature, the system checks the sum of the digits of the passcode. Write logic to compute the sum of the digits of a given number.

Logic Written Below :

1) Get the numerical passcode as input from the user.
2) Initialize the sum equal to 0
3) Iterate through for loop.
4) Add digit to sum and print the sum of digits

```python
# Method - 1 : Without hardcoding the value, getting the input from the user and written the code.
passcodes = input("Enter the Passcodes: ")
sum = 0
for each_digits in passcodes:
    sum = sum + int(each_digits)
print(sum)
```

```python
# Method - 2 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def digilocker():
    passcodes = input("Enter the Amount: ")
    sum = 0
    for each_digits in passcodes:
        sum = sum + int(each_digits)
    print("The Sum of Digits is : ", sum)
digilocker()
```

```python
# Method - 3 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def digilocker(passcodes):
    sum = 0
    for each_digits in str(passcodes):
        sum = sum + int(each_digits)
    return "The Sum of Digits is : ", sum
digilocker(123456)
```

```python
# Method - 4 : By defining classname, function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the classname.functionname() with no arguments.
class NumericalPasscodes:
    def digilocker():
        passcodes = input("Enter the Amount: ")
        sum = 0
        for each_digits in passcodes:
            sum = sum + int(each_digits)
        print("The Sum of Digits is : ", sum)
NumericalPasscodes.digilocker()
```
