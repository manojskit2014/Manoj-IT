print Odd Numbers in the list
(20,10,16,19,25,1,276,188)
19 is odd
25 is odd
1 is odd

```python
list = [20,10,16,19,25,1,276,188]
for i in list:
    if (i %2 != 0):
        print(i, "is odd")
```

(20,10,16,19,25,1,276,188)
20 is odd
10 is odd
16 is odd
276 is odd
188 is odd

```python
list = [20,10,16,19,25,1,276,188]
for i in list:
    if (i %2 == 0):
        print(i, "is even")
```
