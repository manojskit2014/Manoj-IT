Scenario 7 : A retail store offers a 20% discount if a customer’s total order exceeds $100. Write logic to calculate the final amount to be paid after applying the discount.

Logic Written Below :

1) Get the input from the user for the variable total_order.
2) If the total_order exceeds $100.
3) Create one variable discount and calculate the total_amount - (total_amount * (disocunt / percentage)) and store it in the variable discount.
4) User pay the final amount after discount.

```python
# Method - 1 : Hardcoding the total_order 110 and calculate the discounts
total_order = 110
if total_order >= 100:
    total_amount = total_order - (total_order * (20 / 100))
    print("After discount the amount to pay is : ", total_amount)
else:
    print("No discount the amount to pay is : ", total_order)
```

```python
# Method - 1a : Hardcoding the total_order 90 and calculate the discounts
total_order = 90
if total_order >= 100:
    total_amount = total_order - (total_order * (20 / 100))
    print("After discount the amount to pay is : ", total_amount)
else:
    print("No discount the amount to pay is : ", total_order)
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
total_order = int(input("Enter the Amount : "))
if total_order >= 100:
    total_amount = total_order - (total_order * (20 / 100))
    print("After discount the amount to pay is : ", total_amount)
else:
    print("No discount the amount to pay is : ", total_order)
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def totalAmount():
    total_order = int(input("Enter the Amount : "))
    if total_order >= 100:
        total_amount = total_order - (total_order * (20 / 100))
        print("After discount the amount to pay is : ", total_amount)
    else:
        print("No discount the amount to pay is : ", total_order)
totalAmount()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def discountAmount(total_order):
    if total_order >= 100:
        total_amount = total_order - (total_order * (20 / 100))
        return "After discount the amount to pay is : ", total_amount
    else:
        return "No discount the amount to pay is : ", total_order
discountAmount(110)
```
