Scenario 5 : A text-processing tool reverses a given word or sentence for formatting purposes. Write logic to take a word or sentence as input and produce its reversed version.

Logic Written Below :

1) Read the word or sentence from the user.
2) Convert into list with individual characters.
3) Reverse the individual charactes.
4) Join the reversed charactes into string.
5) print the reveresed characters in output console.

```python
# Method - 1 : Hardcoding the string values and converting into list with individual characters
strings = 'Information'
characters = list(strings)
print("The Individual Characters in List is : ", characters)
ch = characters[::-1]
print("The Reversed Individual Characters in List is : ", ch)
reversed_text = "".join(ch)
print("Reversed Text is : ", reversed_text)
```

```python
# Method - 2 : Without hardcoding the value, getting the input from the user and written the code.
names = eval(input("Enter the Number : "))
characters = list(names)
print("The Individual Characters in List is : ", characters)
ch = characters[::-1]
print("The Reversed Individual Characters in List is : ", ch)
reversed_text = "".join(ch)
print("Reversed Text is : ", reversed_text)
```

```python
# Method - 3 : By defining function with no arguments get the input from the user and written the code with print() function and 
# no return statement used and calling the functionname() with no arguments.
def ListReverseString():
    names = eval(input("Enter the String: "))
    characters = list(names)
    print("The Individual Characters in List is : ", characters)
    ch = characters[::-1]
    print("The Reversed Individual Characters in List is : ", ch)
    reversed_text = "".join(ch)
    print("Reversed Text is : ", reversed_text)
ListReverseString()
```

```python
# Method - 4 : By defining function with arguments I am not getting the input from the user and written the code without print() function and 
# return statement used and calling the functionname() with arguments.
def ListReverseString(names):
    characters = list(names)
    print("The Individual Characters in List is : ", characters)
    ch = characters[::-1]
    print("The Reversed Individual Characters in List is : ", ch)
    reversed_text = "".join(ch)
    return "Reversed Text is : ", reversed_text
ListReverseString("Information")
```
